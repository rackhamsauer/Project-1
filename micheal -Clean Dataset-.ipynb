```python
# Dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
# CSV files
dataset_one = Path("Datasets/10000_movies(1915-2023).csv")
dataset_two = Path("Datasets/top_1000_popular_movies_tmdb.csv")
```

```python
# Convert to dataframes
one_df = pd.read_csv(dataset_one)
two_df = pd.read_csv(dataset_two, lineterminator='\n')
```

```python
one_df
```

```python
# Delete unneeded columns
del one_df["Unnamed: 0"]
del one_df["Stars"]
del one_df["Run Time in minutes"]
del one_df["Description"]
del one_df["MetaScore"]
one_df
```

```python
two_df
```

```python
# Delete unneeded columns
del two_df["Unnamed: 0"]
del two_df["id"]
del two_df["tagline"]
del two_df["release_date"]
del two_df["original_language"]
del two_df["popularity"]
del two_df["overview"]
del two_df["runtime"]
two_df
```

```python
# Rename columns for merging
two_df = two_df.rename(columns={"title": "Movie Name"})
two_df
```

```python
merged_df = pd.merge(two_df, one_df, on="Movie Name", how="left")
merged_df
```

```python
# Drop missing data
del merged_df["budget"]
clean_df = merged_df.dropna(how="any")
clean_df
```

```python
# Replace empty brackets
clean_df["production_companies"] = clean_df["production_companies"].replace({"[]": "*"})
clean_df.drop(clean_df[clean_df["production_companies"]=="*"].index)
clean_df
```

```python
# Drop duplicates
clean_df = clean_df.drop_duplicates(subset="Movie Name")
clean_df
```

```python
clean_df = clean_df.rename(columns={"vote_average": "Vote Average",
                                   "vote_count": "Vote Count",
                                   "production_companies": "Production Companies",
                                   "revenue": "Revenue"})
del clean_df["genres"]
clean_df
```

```python
# Save csv file
clean_df.to_csv("Datasets/clean_data.csv", index=False, header=True)
```

```python
clean_data = Path("Datasets/clean_data.csv")
clean_data_df = pd.read_csv(clean_data, lineterminator='\n')
clean_data_df
```

```python
filtered_df = clean_data_df.loc[(clean_data_df['Year of Release'] > 2009)
                     & (clean_data_df['Year of Release'] < 2020)]
filtered_df
```

```python
# Save filtered csv file
filtered_df.to_csv("Datasets/filtered_data.csv", index=False, header=True)
```

```python
average_ratings_by_genre = filtered_df.groupby('Genre')['Movie Rating'].mean()
average_ratings_by_genre.sort_values()
```

```python
average_ratings_by_genre.plot(kind="bar", figsize=(20,3.5), title = "Average Ratings by Genre")

# Set a title for the chart


plt.show()
plt.tight_layout()
```
